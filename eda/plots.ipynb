# Plots

Below we use the final_data.csv file, that was created from the pipeline.

## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/home/mark/Documents/interviews/tennis/data/final_data.csv")
```

```python
data
```

Surface

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
ax = sns.countplot(x="surface", data=data)
plt.title("Surface")
plt.show()
```

Minutes

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="best_of", y="minutes", data=data)
plt.title("Minutes")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="best_of", y="minutes", hue="surface", data=data)
plt.title("Minutes")
plt.show()
```

Match Completion % (by surface vs total count)

```python
match_completion = data.groupby(["surface", "match_completed"]).size().reset_index(name="count")
match_completion["percentage"] = match_completion.groupby("surface")["count"].transform(lambda x: x/x.sum())
match_completion
```

Age

```python
plt.figure(figsize=(10,6))
sns.histplot(data, x="player_age", hue="win", bins=20, kde=True)
plt.title("Age of Players")
plt.show()
```
